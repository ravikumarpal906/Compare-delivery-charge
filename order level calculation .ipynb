```python
#adding new column "Total weight as per X (KG)" from "OrderLevelCalc" in "BillASPerX"
BillASPerX = pd.merge(BillASPerX , OrderLevelCalc[['Order ID' , 'Total weight as per X (KG)']],on='Order ID' , how='left')
```

```python
BillASPerX
```

```python
#create a custom function that value column  "Type of Shipment, Weight Slabs, Forward Fixed Charge,Forward Additional Weight Slab Charge,	RTO Fixed Charge,
#RTO Additional Weight Slab, Charge	Total weight as per X (KG)" row wise and return Expected Charge as per company X (Rs.)



def Expected_Charge_as_per_X(row):
  if row['Type of Shipment'] == 'Forward charges':
    if (row['Total weight as per X (KG)'] % row['Weight Slabs'])!=0:
      return (row['Forward Fixed Charge'] +( row['Forward Additional Weight Slab Charge']*(row['Total weight as per X (KG)'] // row['Weight Slabs'])))
    else:
      return (row['Forward Fixed Charge'] + (row['Forward Additional Weight Slab Charge']*((row['Total weight as per X (KG)'] // row['Weight Slabs'])-1)))
  else:
    if (row['Total weight as per X (KG)'] % row['Weight Slabs'])!=0:
      return (row['Forward Fixed Charge'] +row['RTO Fixed Charge']+( (row['Forward Additional Weight Slab Charge'] + row['RTO Additional Weight Slab Charge'])*(row['Total weight as per X (KG)'] // row['Weight Slabs'])))
    else:
      return (row['Forward Fixed Charge'] +row['RTO Fixed Charge'] + ((row['Forward Additional Weight Slab Charge'] + row['RTO Additional Weight Slab Charge'])*((row['Total weight as per X (KG)'] // row['Weight Slabs'])-1)))
```

```python
#call the custom "Expected_Charge_as_per_X" function
BillASPerX["Expected Charge as per X (Rs.)"] = BillASPerX.apply(Expected_Charge_as_per_X, axis=1)
```

```python
BillASPerX["Expected Charge as per X (Rs.)"] = BillASPerX["Expected Charge as per X (Rs.)"].round(2)
```

```python
# add new column 'Difference Between Expected Charges and Billed Charges (Rs.)' by substracting from 'Expected Charge as per X (Rs.)' by 'Charges Billed by Courier Company (Rs.)'
OrderLevelCalc['Difference Between Expected Charges and Billed Charges (Rs.)'] =OrderLevelCalc['Expected Charge as per X (Rs.)'] - OrderLevelCalc['Charges Billed by Courier Company (Rs.)']
```

```python
#create new dataframe thet store number of order have Expected Charge as per X (Rs.)	and Charges Billed by Courier Company (Rs.) are same and sum of total order that is correctly charged
new_df = OrderLevelCalc[OrderLevelCalc['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0]
total_charges = new_df['Charges Billed by Courier Company (Rs.)'].sum()
Summary = pd.DataFrame({'Count': [new_df.shape[0]], 'Amount': [total_charges]},index=pd.Index(['Total orders where X has been correctly charged']))
Summary
```

```python
#create new row in  data frame "Summary" thet store number of order that Expected Charge as per X (Rs.)	is lower then Charges Billed by Courier Company (Rs.)  and sum of Total Orders where X has been overcharged

new_df = OrderLevelCalc[OrderLevelCalc['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0]
total_charges = abs(new_df['Difference Between Expected Charges and Billed Charges (Rs.)'].sum())
new_df3 = pd.DataFrame({'Count': [new_df.shape[0]], 'Amount': [total_charges]},index=pd.Index(['Total Orders where X has been overcharged']))
Summary = pd.concat([Summary, new_df3])
Summary
```

```python
#create new row in  data frame "Summary" thet store number of order that Expected Charge as per X (Rs.)	is greater then Charges Billed by Courier Company (Rs.)  and sum of Total Orders where X has been undercharged

new_df = OrderLevelCalc[OrderLevelCalc['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0]
total_charges = new_df['Difference Between Expected Charges and Billed Charges (Rs.)'].sum()
new_df4 = pd.DataFrame({'Count': [new_df.shape[0]], 'Amount': [total_charges]},index=pd.Index(['Total Orders where X has been undercharged']))
Summary = pd.concat([Summary, new_df4])
Summary
```
